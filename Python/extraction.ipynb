```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# file_path = "../Data/chr1.27120000.27895000/hic.clean.1/hic.clean.csv.gz"
file_path = "../Data/chr2.230805000.231690000/hic.clean.1/hic.clean.csv.gz"
df = pd.read_csv(file_path)

df
```

```python
pivot_table = df.pivot(index="i1", columns="j1", values="fq")

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, fmt=".2f", cmap="YlOrBr", cbar_kws={'label': 'quantile normalized frequency(fq)'})
plt.title('Quantile Normalized Frequency(fq)')
plt.xlabel('j1')
plt.ylabel('i1')
plt.show()
```
